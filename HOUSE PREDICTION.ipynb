```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("housing.csv")
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("median_house_value", axis=1)
```

```python
y = df["median_house_value"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist()
```

```python
train_data["ocean_proximity"]

pd.get_dummies(train_data["ocean_proximity"]).astype(int)
```

```python
train_data = pd.concat([train_data.drop("ocean_proximity", axis=1), pd.get_dummies(train_data["ocean_proximity"]).astype(int)], axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data.hist()
```

```python
train_data["total_rooms"] = np.log(train_data["total_rooms"] +1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] +1)
train_data["population"] = np.log(train_data["population"] +1)
train_data["households"] = np.log(train_data["households"] +1)
train_data["INLAND"] = np.log(train_data["INLAND"] +1)
train_data["ISLAND"] = np.log(train_data["ISLAND"] +1)
```

```python
train_data.hist(figsize=(12, 8))
```

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x="latitude", y="longitude", data= train_data, hue="median_house_value", cmap="coolwarm")
```

```python
train_data["bedroom_ratio"] = train_data["total_bedrooms"] / train_data["total_rooms"]
train_data["household_rooms"] = train_data["total_rooms"] / train_data["households"]
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression


x_train = train_data.drop(["median_house_value"], axis=1)
y_train = train_data["median_house_value"]

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

pd.get_dummies(test_data["ocean_proximity"]).astype(int)

test_data = pd.concat([test_data.drop("ocean_proximity", axis=1), pd.get_dummies(test_data["ocean_proximity"]).astype(int)], axis=1)

test_data["total_rooms"] = np.log(test_data["total_rooms"] +1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] +1)
test_data["population"] = np.log(test_data["population"] +1)
test_data["households"] = np.log(test_data["households"] +1)
```

```python
test_data["ISLAND"] = np.log(test_data["ISLAND"] +1)

test_data["bedroom_ratio"] = test_data["total_bedrooms"] / test_data["total_rooms"]
test_data["household_rooms"] = test_data["total_rooms"] / test_data["households"]
```

```python
test_data
```

```python
x_test, y_test = test_data.drop("median_house_value", axis=1), test_data["median_house_value"]
```

```python
reg.score(x_test, y_test)
```

```python
#Random Forest

from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [3, 10, 30],
    "max_features": [2, 4, 6, 8]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                          scoring="neg_mean_squared_error",
                          return_train_score=True)

grid_search.fit(x_train, y_train)
```

```python
best_forest =grid_search.best_estimator_
```

```python
best_forest.score(x_test, y_test)
```
